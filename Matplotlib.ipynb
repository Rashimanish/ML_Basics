```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Plotting the data

```python
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis and y-axis lables
plt.plot(x,y)
plt.xlabel('angel')
plt.ylabel('sinvalue')
plt.title('sin wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['english', 'french', 'spanish', 'latin','german']
people = [100,50,150,40,70]
ax.bar(languages , people )
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.title('Language Dispersion')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['english', 'french', 'spanish', 'latin','german']
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
